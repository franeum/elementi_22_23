```python
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
THRESHOLD = 0.00002
```

```python
x = np.linspace(-100, 0, 100)
```

```python
def dbspl(n):
    return 20 * math.log10(n/THRESHOLD)
```

```python
# V2/V1 = 10^(A/20)

def dbtopow(db):
    return 10**(db/10)
```

```python
y = np.array(list(map(dbtopow, x)))
```

```python
y
```

```python
plt.plot(x,y)
```

```python
pd.DataFrame(np.array([x,y]).T).to_csv("./plot01.csv", sep=' ', index=None, header=False)
```
