```python
# importing the required module
import matplotlib.pyplot as plt
import math
```

```python
# x axis values
x = list(range(1000))
# corresponding y axis values
y = [math.sin(v/100*(2*math.pi)) for v in x]
  
# plotting the points 
plt.plot(x, y)
```

```python
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('My first graph!')
```
